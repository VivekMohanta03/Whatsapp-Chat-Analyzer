```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Students Pro Max.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{1,2}\s(?:am|pm)\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
from datetime import datetime
def convert_to_24_hour_format(dates):
    dt_obj = datetime.strptime(dates, '%d/%m/%Y, %I:%M%p - ')
    return dt_obj.strftime('%d/%m/%Y, %H:%M - ')

dates = [convert_to_24_hour_format(ts.replace('\u202f', '')) for ts in dates]

print(dates)
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date': dates})

#converting message data type

df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'message_date': 'date'}, inplace=True)

df.tail()
```

```python
df.shape
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head()

```

```python
#extracting the year
df['year'] = df['date'].dt.year

df.head()
```

```python
#extracting the month name
df['month'] = df['date'].dt.month_name()

df.head()
```

```python
#extracting the day
df['day'] = df['date'].dt.day

df.head()
```

```python
#Extracting hour and minutes
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute

df.head()
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
